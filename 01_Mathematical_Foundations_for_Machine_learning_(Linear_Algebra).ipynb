# Linear algebra

Linear algebra is the study of vectors, matrices, and operations on them. It forms the basis of many machine learning algorithms, especially those involving high-dimensional data and neural networks.

## What are vector spaces though?

A vector space is a place where you can play with directions and magnitudes in a way that follows specific rules, like adding them together or scaling them. These rules are what make it a "space" where lots of mathematical operations can happen.

The coordinate plane is an example of a vector space. It is a 2D space $\mathbb{R}^2$ where each point is identified by a pair of real numbers (x,y). Note that vector spaces generalize to spaces of higher dimensions $\mathbb{R}^d$


## How do you represent a dataframe in mathematics

It is represented using a numerical matrix. Each row in the df corresponds to a row in the matrix and each column to a column in the matrix.

## How to represent a datapoint in mathematics?

**A data point is known as a dataframe row (==matrix row). Each row in the matrix, represents a vector, so we can say that a vector is a mathematical representation of a data point**.

A vector is therefore a mathematical object that represents a quantity that has both magnitude and direction. In the context of linear algebra, a vector is often represented as an ordered collection of numbers. The numbers in this collection reflect the cordinates of the vector in a specific cordinate system.

A. Row vector

This is a vector represented as a 1 x d matrix, where d = number of components

$$ \mathbf{a} = \begin{bmatrix} a_1 & a_2 & a_3 \end{bmatrix} $$


B. Column vector

Represented as an d x 1 matrix.

$$ \mathbf{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$

## Properties of a vector

A. Dimensionality

This is the number of columns (aka features) it has. For example, vector "a" below has 3 columns:

$$ \vec{\mathbf{a}} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$

B. Magnitude of a vector

**This is also known as the Norm or Length of a vector.** Note the following:

- The magnitude of a vector represents the size or extent of the vector. It is always a positive value.

- The magnitude of a vector in a 2D space with two componenets $\ a_1$ and $\ a_2$ is calculated using the Euclidean norm/length formula


$$ \|\mathbf{a}\| = \sqrt{a_1^2 + a_2^2} $$

- For a 3D space, a vector with three components  $\ a_1$, $\ a_2$ and $\ a_3$, the magnitude is calculated as:

$$ \|\mathbf{a}\| = \sqrt{a_1^2 + a_2^2 + a_3^2} $$

- In general, for a D-dimensional vector $\mathbf{a} = [a_1, a_2, ...,a_d]$

$$ \|\mathbf{a}\| = \sqrt{a_1^2 + a_2^2 + \cdots + a_d^2} $$

C. Direction of a vector

- Describes the way the vector points in space
- The direction of a vector can be expressed as a unit vector. **A unit vector simply represents a vector with the same direction but a magnitude of 1.** Why? Unit vectors are useful to allow you to work with direction alone without the complication of varying magnitudes. So typcally you are saying, let's focus on the direction of this path and agree that we'll measure it as a distance of 1.

- The process of converting an original vector to a unit vector is known as normalization. It involves dividing each of its compoonents by its magnitude. The formula is given as:

$$ \hat{\mathbf{a}} = \frac{\mathbf{a}}{\|\mathbf{a}\|} $$

```python
print('Dimensionality:', vec1.shape)
print()

#calculating the magnitude
print('Magnitude of vec1:', np.linalg.norm(vec1))
print('Magnitude of vec2:', np.linalg.norm(vec2))
print('Magnitude of vec3:', np.linalg.norm(vec3))

#where np.linalg.norm is a numpy function created to calculate the magnitude of a vector

#calculating the unit vector

print()
print('Unit vector of vec1:', vec1/np.linalg.norm(vec1))
print('Unit vector of vec2:', vec2/np.linalg.norm(vec2))
print('Unit vector of vec3:', vec3/np.linalg.norm(vec3))
```

## Vector algebra

A. Addition

When two vectors of the same dimension are added, the result will also be in the same dimension. Here, each component is the sum of the corresponding elements of the original vectors.


$$ \mathbf{u} = \begin{bmatrix} u_1 \\ u_2 \\ u_3 \end{bmatrix}, \quad
\mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} $$

$$ \mathbf{u} + \mathbf{v} = \begin{bmatrix} u_1 + v_1 \\ u_2 + v_2 \\ u_3 + v_3 \end{bmatrix} $$

B. Subtraction

The same concept applies as above.

C. Dot Product (AKA Scalar product)

The dot product of two vectors of the same dimension is a scalar value calculated by summing the products of corresponding components. It can be expressed as:

$$ \vec{\mathbf{a}} \cdot \vec{\mathbf{b}} = \sum_{i=1}^{n} a_i b_i $$

## Similarity between two vectors - Distance vs Similarity

**These two concepts are used to quantify the relationship between datapoints and while they are related, they represent opposite aspects of this relationship. E.g., a high similarity corresponds to a low distance between data points, and vice versa.**

Note that the choice between similarity and distance depends on the specific context and requirements of the problem at hand. Also, the choice of a distance metric depends on the specific characteristics of your data and the requirements of your application.

A. Manhattan distance (L1 Norm)

Represents the sum of the absolute differences between the cordinates. It is suitable for scenarios where movement can only occur along grid lines.

$$\begin{equation*}
distance(\vec{\mathbf{a}}, \vec{\mathbf{b}}) = \sum_{i=1}^{d}|a_i - b_i|
\end{equation*}$$


$$\ D = |X_2-X_1| + |Y_2-Y_1|$$

B. Euclidean distance (L2 Norm)

Measures the straight-line distance between two points in the euclidean space. It is usually used for continuous data. This is computationaly expensive.

$$\begin{equation*}
distance(\vec{\mathbf{a}}, \vec{\mathbf{b}}) = \sqrt{\sum_{i=1}^{d}(a_i - b_i)^2}
\end{equation*}$$

The easiest way to remember this formula is the formula for calculating the hypothenus according to pythagoras theorem as seen below:

$$\ D = \sqrt{(X_2-X_1)^2 + (Y_2-Y_1)^2}$$

C. Minkowski distance (Lp Norm)

It is a generalization of both distance metric mentioned above. The p parameter allows us to vary between the two

$$\begin{equation*}
distance(\vec{\mathbf{a}}, \vec{\mathbf{b}}) = \left(\sum_{i=1}^{d}|a_i - b_i|^p\right)^{1/p}
\end{equation*}$$

```python
##manhattan distance

from sklearn.metrics import pairwise
md1 = pairwise.manhattan_distances([vec1], [vec2])
md2 = pairwise.manhattan_distances([vec1, vec2, vec3])

print('Manhattan distance between Vec1 and Vec2 is: ', md1)
print()
print('Manhattan distance between the three vectors:', md2)
```

```python
##euclidean distance

from sklearn.metrics import pairwise
ed1 = pairwise.euclidean_distances([vec1, vec2])
ed2 = pairwise.euclidean_distances([vec1, vec2, vec3])

print('Euclidean distance between vector 1 and 2:', ed1)
print('Euclidean distance between the three vectors:', ed2)
```

## Cosine similarity (represented as $\cos\theta$)

The cosine similarity is concerned with measuring the cosine of the angle between two vectors. It is often used for text and high-dimensional data where the magnitude of the vectors is not important. **Cosine similarity ranges from -1 to 1.**

$$\begin{equation*}
\cos\theta_{a,b} = \frac{\vec{\mathbf{a}} \cdot \vec{\mathbf{b}}}{\|\vec{\mathbf{a}}\| \times \|\vec{\mathbf{b}}\|}
\end{equation*}$$


## Cosine distance (represented as $1 - \cos\theta$)

It measures dissimilarity and shows how far apart two vectors are in terms of orientation. A smaller cosine distance indicates greater similarity. **Cosine distance ranges from 0 to 2**

$$distance(a,b) = 1 - \cos\theta_{a,b}$$


## Angular distance (represented as $\theta$)

This is also known as angular seperation or angular similarity. It is a measure of the seperation between two vectors or points in a multi-dimensional space. In simple terms, it measures the angle of sepration between two vectors. It helps with quantifying the difference in orientation or direction between two entities. It is the inverse of the cosine similarity

$$\begin{equation*}
\theta_{a,b} = \cos^{-1} \left(\frac{\vec{\mathbf{a}} \cdot \vec{\mathbf{b}}}{\|\vec{\mathbf{a}}\| \times \|\vec{\mathbf{b}}\|} \right)
\end{equation*}$$

```python
##cosine similarity

from sklearn.metrics import pairwise

cs1 = pairwise.cosine_similarity([vec1, vec2])
cs2 = pairwise.cosine_similarity([vec1, vec2, vec3])

print('Cosine similarity between vector 1 and 2:', cs1)
print()
print('Cosine similarity between vector 1, 2,3:', cs2)
```

```python
##cosine distance

from sklearn.metrics import pairwise

cd1 = pairwise.cosine_distances([vec1, vec2])
cd2 = pairwise.cosine_distances([vec1, vec2, vec3])

print('Cosine distance between vector 1 and 2\n', cd1)
print()
print('Cosine distance between vectors 1, 2, 3:\n', cd2)
```

```python
# Angular distances

from sklearn.metrics import pairwise

cos_sim1 = pairwise.cosine_similarity([vec1, vec2])
ad1 = np.degrees(np.arccos(cos_sim1))
print('Angular distance between vector 1 and 2:\n', ad1)
print()

cos_sim2 = pairwise.cosine_similarity([vec1, vec2, vec3])
ad2 = np.degrees(np.arccos(cos_sim2))
print('Angular distance between vector 1, 2, 3:\n', cos_sim2)
```

The error arises because the arccos function in NumPy is receiving an invalid input. For cosine similarity, the value must lie in the range [−1,1], as the cosine of an angle cannot exceed these bounds. If the computed cosine similarity values are slightly outside this range due to numerical precision issues, the arccos function will throw a warning.

To fix this issue we clamp the Cosine Similarity values. This ensure the cosine similarity values are clamped to the range [−1,1] before passing them to arccos.

```python
# Angular distances
cos_sim1 = pairwise.cosine_similarity([vec1, vec2])
clamped_cos_sim1 = np.clip(cos_sim1, -1, 1)
ad1 = np.degrees(np.arccos(clamped_cos_sim1))
print('Angular distance between vector 1 and 2:\n', ad1)
print()

cos_sim2 = pairwise.cosine_similarity([vec1, vec2, vec3])
clamped_cos_sim2 = np.clip(cos_sim2, -1, 1)
ad2 = np.degrees(np.arccos(clamped_cos_sim2))
print('Angular distance between vector 1, 2, 3:\n', ad2)
```

## Dot product

It is used to measure the similarity between two vectors. Some properties of dot products include:

A. 2 vectors are perpendicular if their dot products is zero

B. The dot product of a vector with itself is equal to the square of its magnitude.

$$\begin{equation*}
\vec{\mathbf{a}} \cdot \vec{\mathbf{a}} = \|\vec{\mathbf{a}}\|^2
\end{equation*}$$

C. When the vectors are normalized (that is, have unit length), the dot product is equivalent to the cosine of the angle between them.

$$\begin{equation*}
\cos\theta_{a,b} = \frac{\vec{\mathbf{a}} \cdot \vec{\mathbf{b}}}{\|\vec{\mathbf{a}}\| \times \|\vec{\mathbf{b}}\|}
\end{equation*}$$

The formula for dot product comes in various forms:

A. Method 1

$$Dot \ product = np.dot(vec1, vec2)$$


B. Method 2 -- where magnitude is calculated using np.linalg.norm
$$Dot \ product =||vec_1|| * ||vec_2|| * \cos(\theta) $$

Where $\cos(\theta)$ is the cosine angle between vectors


C. Method 3 -- where magnitude is calculated using np.euclidean_distances

$$Dot \ product =||vec_1|| * ||vec_2|| * \cos(\theta) $$


```python
## Method 1 of finding the dot product

print(np.dot(vec1, vec2))
```

```python
## Method 2

#computing magnitudes for each vector
mag_vec1 = np.linalg.norm(vec1)
mag_vec2 = np.linalg.norm(vec2)

#computing cos theta between vectors
cos_sim = pairwise.cosine_similarity([vec1], [vec2])

print('Dot product:', mag_vec1 * mag_vec2 * cos_sim)
```

```python
##Method 3

#computing magnitudes for each vector
mag_vec1 = pairwise.euclidean_distances([[0,0]], [vec1])
mag_vec2 = pairwise.euclidean_distances([[0,0]], [vec2])

#computing cos theta between vectors
cos_sim = pairwise.cosine_similarity([vec1], [vec2])

print('Dot product:', mag_vec1 * mag_vec2 * cos_sim)
```

### Matrix multiplication
It takes two matrices and produces another matrix.

### Matrix inverse

The inverse of a matrix $A$ denoted as $A^{-1}$ is a matrix that when multiplied by $(A)$ results in the identity matrix $I$. That is:

$$A.A_{-1} = I$$

It is important to note that not all matrices have inverses. Therefore, in order for inverse for a matrix to exist, a matrix must:

- Have the same number of rows and columns
- It's determinant != 0

```python
##matrix multiplication
import numpy as np

A=np.array([[1,2], [3,4]])
B=np.array([[2,6], [9,0]])

#method 1 for matrix multiplication
C = np.dot(A,B)
print('Method 1 result:\n', C)

#method 2
D = np.matmul(A,B)
print('Method 2 result:\n', D)
```

```python
##matrix inverse

A = np.array([[3,5], [9,8]])

#checking the detrminant
print('Determinant:',np.linalg.det(A))

#computing the inverse of the matrix
A_inv = np.linalg.inv(A)

print('\n Inverse of A:\n',A_inv)
```

### Eigenvalues and EigenVectors

They are used to understand linear transformations. Given a square matrix, an eigenvector $v$ and its corresponding eigenvalue $\lambda$ satisfy the equation:

$$Av = \lambda v$$

```python
import numpy as np

#computing eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print('Eigenvalues:', eigenvalues)
print()
print('Eigenvectors:')
for i in range(len(eigenvalues)):
    print()
    print('Eigenvalue:', eigenvalues[i])
    print('Eigenvector:', eigenvectors[:,i])
```
