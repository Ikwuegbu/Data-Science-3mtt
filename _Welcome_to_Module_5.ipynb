## Inferential statistics

Inferential statistics is a branch of statistics that allows us to make predictions, decisions, or generalizations about a population based on data collected from a sample. Since it is often impractical or impossible to gather data from an entire population, inferential statistics uses sample data to infer or estimate population parameters, test hypotheses, and make data-driven decisions.

### Common terminologies, concepts and tools in inferential statistics
A. **Population**

This is the entire set of individuals, items, or data points of interest. _Example_: If you're studying the average height of adults in a city, the population would include every adult in that city.

B. **Sample**

A subset of the population selected for analysis. _Example_: Instead of measuring the height of every adult in the city, you might select 1,000 adults to represent the population.

C. **Parameter**

A numerical characteristic or measure of a population (e.g., mean, median, variance). _Example_: The true average height of all adults in the city is a parameter.

D. **Statistic**

A numerical characteristic or measure calculated from a sample. _Example_: The average height of the 1,000 adults in the sample is a statistic.

E. **Hypothesis Testing**

A method used to determine whether there is enough evidence to reject a null hypothesis about a population parameter. _Example_: Testing if the average height of adults in the city is 5'6'' (null hypothesis).

F. **Null Hypothesis**

The default assumption that there is no effect or no difference. _Example_: $H_0$: The average height of adults in the city is 5'6''.

G. **Alternative Hypothesis**

The hypothesis that contradicts the null hypothesis, indicating an effect or difference. _Example_: $H_a$: The average height of adults in the city is not 5'6''.

H. **P-Value**

The probability of observing the data (or something more extreme) if the null hypothesis is true. _Example_: A p-value of 0.03 indicates a 3% chance that the observed data occurred under the null hypothesis.

I. **Confidence Interval**

A range of values that is likely to contain the population parameter with a certain level of confidence (e.g., 95%). _Example_: A confidence interval for average height might be 5'5'' to 5'7'', with 95% confidence.

J. **Significance Level (Œ±)**

The threshold probability for rejecting the null hypothesis, typically set at 0.05 (5%). _Example_: If Œ±=0.05, results with a p-value below 0.05 lead to rejection of $H_0$.

K. **Type I Error**

Rejecting the null hypothesis when it is actually true (false positive). _Example_: Concluding the average height is not 5'6'' when it actually is.

L. **Type II Error**

Failing to reject the null hypothesis when it is false (false negative). _Example_: Concluding the average height is 5'6'' when it is actually not.

M. **Test Statistic**

A standardized value calculated from sample data used to determine whether to reject $H_0$‚Äã. _Example_: The z-score or t-score in hypothesis testing.

N. **Z-Test**

A hypothesis test used when the sample size is large and the population variance is known. _Example_: Testing whether the average height is 5'6'' using a z-test with a large sample size.

O. **T-Test**

A hypothesis test used when the sample size is small, and the population variance is unknown. _Example_: Comparing the mean height of a small sample to 5'6''.

P. **Chi-Square Test**

A test used to determine if there is a significant association between categorical variables. _Example_: Testing whether gender and favorite color are independent.

Q. **ANOVA (Analysis of Variance)**

A statistical test used to compare means among three or more groups. _Example_: Comparing test scores of students from three different schools.

R. **Degrees of Freedom**

The number of independent values that can vary in a statistical calculation. _Example_: In a t-test with a sample size of 20, the degrees of freedom would be n‚àí1=19n‚àí1=19.

S. **Standard Error**

The standard deviation of the sampling distribution of a statistic. _Example_: The standard error of the mean indicates how much the sample mean is expected to vary from the true population mean.

T. **Central Limit Theorem**

The theorem states that the sampling distribution of the sample mean will be approximately normal if the sample size is sufficiently large, regardless of the population's distribution. _Example_: Averaging test scores from 100 students will yield a normal distribution of sample means.

U. **Confidence Level**

The proportion of confidence intervals that contain the true parameter if the study were repeated multiple times. _Example_: A 95% confidence level means we are 95% confident the interval includes the population mean.

V. **Effect Size**

A measure of the magnitude of the difference between groups or the strength of a relationship. _Example_: Cohen‚Äôs d is used to measure the effect size of differences between means.

W. **Regression Analysis**

A statistical method to model and analyze relationships between variables. _Example_: Predicting house prices based on the number of bedrooms and location.

X. **Correlation Coefficient**

A measure of the strength and direction of the relationship between two variables. _Example_: A correlation coefficient of 0.9 indicates a strong positive relationship.

Y. **Sampling Distribution**

The probability distribution of a statistic obtained from multiple samples. _Example_: The distribution of sample means calculated from 1,000 samples of 50 individuals each.


### Why Use Inferential Statistics?

- It‚Äôs often impossible to study every member of a population (e.g., all humans on Earth).
- It saves time and resources by focusing on a smaller, manageable sample.
- It provides tools to measure the accuracy and reliability of these predictions.

### Examples Inferential Statistics in the Real-world

**Example 1: Predicting Average Height in a City**

Imagine you want to know the average height of adults in your city.
- Problem: Measuring everyone‚Äôs height is unrealistic.
- Solution: You measure the height of 500 randomly chosen adults (a sample) and use this data to infer the average height of all adults in the city (the population).

You can also calculate a confidence interval to express the range where the true average likely falls, e.g., "The average height is between 5‚Äô5‚Äô‚Äô and 5‚Äô7‚Äô‚Äô with 95% confidence."

**Example 2: Testing a New Drug**

A pharmaceutical company is testing a new drug designed to lower blood pressure.
- Problem: Testing the drug on the entire population of people with high blood pressure is impractical.
- Solution: The company selects 1,000 volunteers for a trial (the sample). Based on the results, they infer whether the drug works for the entire population and estimate the average reduction in blood pressure.

**Example 3: Election Polling**

Before an election, a polling agency surveys 2,000 voters to predict which candidate will win.
- Problem: Surveying every voter is impossible.
- Solution: Based on the sample responses, the agency estimates the percentage of votes each candidate will receive in the entire population and declares a winner with a margin of error (e.g., "Candidate A has 52% ¬± 3% support").

**Example 4: Quality Control in a Factory**

A factory produces thousands of light bulbs daily, and the management wants to ensure the bulbs meet quality standards.
- Problem: Testing every single bulb for durability would destroy the entire production.
- Solution: They test a sample of 100 light bulbs and infer whether the entire batch meets quality standards.

**Example 5: Studying Public Opinion**

A researcher wants to know how many people in a country support renewable energy policies.
- Problem: Asking every citizen is infeasible.
- Solution: The researcher surveys a representative sample of 10,000 citizens and uses inferential statistics to estimate public opinion with a margin of error.

### Another way to think of inferential statistics

Instead, you take a small spoonful (a sample) to determine the overall taste (the population‚Äôs characteristics). Inferential statistics uses similar logic: it works with small, manageable portions of data to draw conclusions about the whole.
